### TASK: What % of people from each country knows Python?

```python
data.groupby('Country')['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').value_counts(normalize=True)).unstack()
```

```python
python_countries = data.groupby('Country')['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').value_counts(normalize=False)).unstack()
```

```python
python_countries['Total'] = data['Country'].value_counts()
python_countries.sort_values(by=True, ascending=False, inplace=True)
python_countries.drop('total', axis=1, inplace=True)
python_countries['Percentage_of_Pythoners'] = python_countries[True]/python_countries['Total']
python_countries
```

```python
python_countries['Percentage_of_Pythoners'].dropna().sort_values(ascending=False)
```

```python
python_countries['total'] = data['Country'].value_counts()
python_countries['total'].sort_values(ascending=False)
```

```python
python_countries.nlargest(10, True)
```
